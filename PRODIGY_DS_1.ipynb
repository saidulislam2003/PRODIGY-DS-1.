```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/color.csv')
```

```python
df.head()
```

```python
color = df[["Red", "Blue", "Green"]].sum()
```

```python
color
```

```python
color.plot(kind = "bar", color = ["Red", "Blue", "Green"])
plt.xlabel("Colors")
plt.ylabel("Number of people")
plt.title("Age 20-30 color liking survey");
```
